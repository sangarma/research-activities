```python
# import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.stats.outliers_influence import variance_inflation_factor
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv("data_Q037.csv")
data.describe()
```

```python
data.head(20)
```

```python
columns = data.columns
for col in columns:
    print(col)
```

```python
cat_variable = ['Age(years)','Education', 'Gender', 'Income','Race/Ethnicity']
for var in cat_variable:
    print(var)
```

```python
one_hot_cncoded = pd.get_dummies(data[cat_variable], dummy_na = True)
```

```python
one_hot_cncoded
```

```python
data = data.drop(cat_variable, axis = 1)
```

```python
f_data = data.join(one_hot_cncoded)
```

```python
for var in f_data:
    print(var)
```

```python
features = ['Age(years)_18 - 24', 'Age(years)_25 - 34','Age(years)_35 - 44','Education_High school graduate', 'Education_Less than high school',
       'Education_Some college or technical school', 'Gender_Female', 'Income_$15,000 - $24,999', 'Income_$25,000 - $34,999',
       'Income_$35,000 - $49,999','Income_Less than $15,000', 'Race/Ethnicity_2 or more races','Race/Ethnicity_American Indian/Alaska Native', 'Race/Ethnicity_Asian',
       'Race/Ethnicity_Hispanic','Race/Ethnicity_Non-Hispanic Black']
"""features = ['YearStart','Age(years)_18 - 24', 'Age(years)_25 - 34','Age(years)_35 - 44', 'Age(years)_45 - 54', 'Age(years)_55 - 64',
       'Education_College graduate','Education_High school graduate', 'Education_Less than high school',
       'Education_Some college or technical school', 'Income_$15,000 - $24,999', 'Income_$25,000 - $34,999',
       'Income_$35,000 - $49,999', 'Income_$50,000 - $74,999','Income_$75,000 or greater', 'Income_Data not reported','Income_Less than $15,000', 'Race/Ethnicity_2 or more races','Race/Ethnicity_American Indian/Alaska Native', 'Race/Ethnicity_Asian',
       'Race/Ethnicity_Hawaiian/Pacific Islander', 'Race/Ethnicity_Hispanic','Race/Ethnicity_Non-Hispanic Black','Race/Ethnicity_Non-Hispanic White', 'Race/Ethnicity_Other']"""
X = f_data[features]
```

```python
y = f_data['Data_Value']
```

```python
model = LinearRegression()
model.fit(X,y)
```

```python
vif = pd.DataFrame()
vif["VIF Factor"] = [variance_inflation_factor(X, i) for i in range(X.shape[1])]
vif["features"] = features
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
# Predict the outcome variable for the testing data
y_pred = model.predict(X_test)

# Calculate evaluation metrics
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

# Print the evaluation metrics
print("Mean Absolute Error: {:.2f}".format(mae))
print("Mean Squared Error: {:.2f}".format(mse))
print("Root Mean Squared Error: {:.2f}".format(rmse))
print("R-squared: {:.2f}".format(r2))
```

```python
from sklearn.tree import DecisionTreeRegressor
# Create a decision tree regressor object
regressor = DecisionTreeRegressor(max_depth=100, max_features=3)
#X = data[cat_variable]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=43)
# Train the decision tree model on the training data
regressor = regressor.fit(X_train, y_train)

# Use the trained model to make predictions on the test data
y_pred = regressor.predict(X_test)

# Calculate evaluation metrics
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

# Print the evaluation metrics
print("Mean Absolute Error: {:.2f}".format(mae))
print("Mean Squared Error: {:.2f}".format(mse))
print("Root Mean Squared Error: {:.2f}".format(rmse))
print("R-squared: {:.2f}".format(r2))
# Calculate the mean squared error of the model
mse = mean_squared_error(y_test, y_pred)

print("Mean squared error:", mse)
```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(regressor, X,y, cv=50, scoring='r2')
scores
```

```python
# Import the model we are using
from sklearn.ensemble import RandomForestRegressor
# Instantiate model with 1000 decision trees
rf = RandomForestRegressor(n_estimators = 1000, random_state = 49)
rf.fit(X_train, y_train)
# Use the trained model to make predictions on the test data
y_pred = rf.predict(X_test)

# Calculate evaluation metrics
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

# Print the evaluation metrics
print("Mean Absolute Error: {:.2f}".format(mae))
print("Mean Squared Error: {:.2f}".format(mse))
print("Root Mean Squared Error: {:.2f}".format(rmse))
print("R-squared: {:.4f}".format(r2))
# Calculate the mean squared error of the model
mse = mean_squared_error(y_test, y_pred)

print("Mean squared error:", mse)
```
